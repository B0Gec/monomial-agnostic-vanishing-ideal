```python
%reload_ext autoreload
%autoreload 2

import sys
sys.path.append('..')

import numpy as np
import matplotlib.pyplot as plt

# from mavi.numpy.vanishing_ideal import VanishingIdeal
from mavi.vanishing_ideal import VanishingIdeal
```

```python
theta = [np.pi*i/3 for i in range(6)]
X = np.vstack((np.cos(theta), np.sin(theta))).T

plt.plot(X[:,0], X[:,1], "o")
```

## Gradient normalization

```python
vi = VanishingIdeal()
vi.fit(X, 0.01, method="grad")  # grad normalization
```

```python
len(vi.basis.nonvanishings()) # (maximum deg of nonvanishing polynomials) - 1 
FX = vi.evaluate(X, target="nonvanishing") 
GX = vi.evaluate(X, target="vanishing")
dGX = vi.gradient(X, target='vanishing')
print("Gradient norm\n {}\n".format(np.linalg.norm(dGX, axis=0)))
```

```python
print(np.linalg.norm(GX, axis=0))
vi.plot(X, splitshow=True)
```

## VCA

```python
vi = VanishingIdeal()
vi.fit(X, 0.1, method="vca")  # vca
```

```python
len(vi.basis.nonvanishings()) # (maximum deg of nonvanishing polynomials) - 1 
FX = vi.evaluate(X, target="nonvanishing") 
GX = vi.evaluate(X, target="vanishing")
dGX = vi.gradient(X, target='vanishing')

print("Gradient norm\n {}\n".format(np.linalg.norm(dGX, axis=0)))
print("Non-van. polys are normalized w.r.t. evaluation\n {}\n".format(np.linalg.norm(FX, axis=0)))
```

```python
print(np.linalg.norm(GX, axis=0))
vi.plot(X, target="vanishing", splitshow=True)
```
